# Survey Answers Data Preparation 

### Data Import

```python
# import survey answers
df = pd.read_csv(r"C:\\Users\Chennaf\Documents\Survey.csv" , delimiter =';')
```

```python
df.info()
```

### Data Cleaning

```python
df
```

```python
# Drop useless time column
df = df.drop('Timestamp', 1)
df.head()
```

```python
# Drop first three introductory questions + 'Unnamed:5' empty column
df.drop(df.columns[0], axis=1, inplace=True)
df.drop(df.columns[0], axis=1, inplace=True)
df.drop(df.columns[0], axis=1, inplace=True)
df.drop(df.columns[1], axis=1, inplace=True)

# Drop last 4 questions of the survey - they are for another analysis
df.drop(df.columns[10], axis=1, inplace=True)
df.drop(df.columns[10], axis=1, inplace=True)
df.drop(df.columns[10], axis=1, inplace=True)
df.drop(df.columns[10], axis=1, inplace=True)
df.drop(df.columns[10], axis=1, inplace=True)

# drop first 4 rows that were only used for trial
df.drop(df.index[[0,1,2,3,4]], inplace=True)
df = df.reset_index(drop=True)

# dropping any extra rows with Null values in them
df = df.dropna(how='any',axis=0) 

df.head()
```

```python
# rename expertise level column
df.rename(columns={ df.columns[10]: "Profile" }, inplace = True)

# renaming profiles
df.loc[df['Profile'].str.contains('End', case=False), 'Profile'] = 'End_user'
df.loc[df['Profile'].str.contains('Senior', case=False), 'Profile'] = 'Expert'
df.loc[df['Profile'].str.contains('Junior', case=False), 'Profile'] = 'Junior'
df.head()
```

```python
# grouping all answers according to the profiles
df = df.groupby(['Profile']).mean()
df.head()
```

```python
# renaming columns questions with their indices
df.columns = np.arange(len(df.columns))
df.head()
```

```python
# add mean value of all profiles combines
df.loc['mean_all'] = df.mean()
df
```

```python
# creating 3 seperate series - one for each profile
end = df.iloc[0]
exp = df.iloc[1]
jun = df.iloc[2]
allmean =df.iloc[3]

print(jun)
```

# Features - Data preperation 

```python
# import csv of chosen questions for the survey that was prepared in advance 
chosen = pd.read_csv(r"C:\\Users\Chennaf\Documents\Chosen.csv" , delimiter =';')
chosen.head()
```

```python
# drop useless columns cluster class + name + question #
chosen = chosen.drop('Cluster Class', 1)
chosen = chosen.drop('Name', 1)
chosen = chosen.drop('Question', 1)
chosen.head()
```

# Combining results

```python
# creating three dataframes, one for each user type
end_df = chosen.copy()
exp_df = chosen.copy()
jun_df = chosen.copy()
mean_all = chosen.copy()

jun_df.head()
```

```python
# multiplying each dataframe of each profile by the average rating given in survey
result_end = end_df.mul(end, axis=0)
result_exp = exp_df.mul(exp, axis=0)
result_jun = jun_df.mul(jun, axis=0)
result_all = mean_all.mul(allmean, axis=0)
result_all
```

```python
# adding mean row to each dataframe
result_end.loc['mean'] = result_end.mean()
result_exp.loc['mean'] = result_exp.mean()
result_jun.loc['mean'] = result_jun.mean()
result_all.loc['mean'] = result_all.mean()

# adding average row for all values in each column and creating seperate dataframe for means only
#mean_end = result_end.mean()
#mean_exp = result_exp.mean()
#mean_jun = result_jun.mean()

result_end
```

```python
result_all
```

```python
# concat mean rows from each dataframes into one for a baseline
result = pd.DataFrame()
result = result.append(result_end.loc['mean'], ignore_index=True)
result = result.append(result_exp.loc['mean'], ignore_index=True)
result = result.append(result_jun.loc['mean'], ignore_index=True)
result = result.append(result_all.loc['mean'], ignore_index=True)

# end_user mean is index 0, junior mean is at index 2, expert mean is at index 1 (alphabetical order)
result
```

```python
# splitting dataframe between features and explanation output
feat = result.filter(regex='F_')
out = result.filter(regex='Out_')
feat
```
